```python
import pandas as pd

def read_all_sheets_from_excel(file_path):
    # 步骤2: 使用ExcelFile加载xlsx文件
    xls = pd.ExcelFile(file_path)
    
    # 步骤3: 遍历所有工作表名称，并读取每个工作表到一个DataFrame
    dataframes = {sheet_name: xls.parse(sheet_name) for sheet_name in xls.sheet_names}
    
    # 返回一个字典，键是工作表名称，值是对应的DataFrame
    return dataframes

# 使用函数读取文件
file_path = './Redditors_and_posts_batch_I_anonymized.xlsx'  # 替换为你的文件路径
dataframes = read_all_sheets_from_excel(file_path)

# 现在可以访问`dataframes`字典中的每个DataFrame了
for sheet_name, df in dataframes.items():
    print(f"Sheet name: {sheet_name}\n", df.head())  # 打印每个工作表的前几行

```

```python
import pandas as pd

def read_xlsx_and_count_categories(file_path):
    # 读取.xlsx文件的每个工作表
    xls = pd.ExcelFile(file_path)

    # 初始化一个字典来存储所有类别的计数
    category_counts = {}

    for sheet_name in xls.sheet_names:
        # 读取当前工作表
        df = pd.read_excel(xls, sheet_name)
        
        # 检查"Suicide Risk"列是否存在
        if "Suicide Risk" in df.columns:
            # 统计当前工作表中每个类别的出现次数
            counts = df["Suicide Risk"].value_counts().to_dict()
            # print(counts)
            
            # 合并当前工作表的统计结果到总的计数中
            for category, count in counts.items():
                if category in category_counts:
                    category_counts[category] += count
                else:
                    category_counts[category] = count

    # 返回最终的统计结果
    return category_counts

# 替换为你的文件路径
file_path = './Redditors_and_posts_batch_II_anonymized.xlsx'
category_counts = read_xlsx_and_count_categories(file_path)
print(category_counts)
```

```python
import pandas as pd

# 假设文件路径为 'example.xlsx'，请根据实际情况替换
file_path = './Redditors_and_posts_batch_I_anonymized.xlsx'  # 请将此路径替换为你的实际文件路径

# 读取.xlsx文件的每个工作表
xls = pd.ExcelFile(file_path)

# 初始化一个字典来存储所有类别的计数，不区分大小写
category_counts = {}

for sheet_name in xls.sheet_names:
    # 读取当前工作表
    df = pd.read_excel(xls, sheet_name)
    
    # 检查"Suicide Risk"列是否存在，并转换为小写
    if "Suicide Risk" in df.columns:
        df["Suicide Risk"] = df["Suicide Risk"].str.lower()  # 转换为小写
        counts = df["Suicide Risk"].value_counts().to_dict()
        
        # 合并当前工作表的统计结果到总的计数中
        for category, count in counts.items():
            if category in category_counts:
                category_counts[category] += count
            else:
                category_counts[category] = count

# 输出最终的统计结果
print(category_counts)
```
